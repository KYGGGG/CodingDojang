## SECTION 02 배열 다루기

### 배열의 데이터형 확인과 변형

```python
import numpy as np

nda = np.array([0,1,2,3,4,5])
print(nda, type(nda))
print(nda.dtype)
nda
```

```python
a1 = np.array(['1.2', '2.3', '3.4', '4.5'])
print(a1, a1.dtype)

a2 = a1.astype('float32')
print(a2, a2.dtype)

a3 = a2.astype('int32')
print(a3, a3.dtype)

a3 = a1.astype('int32')
print(a3, a3.dttype)
```

### 배열의 형태 변경

```python
m1 = np.arange(6)
print(m1)
print(m1.shape)
```

```python
m2 = m1.reshape(6, 1)
print(m2, m2.shape)
```

```python
m2 = m1.reshape(2, 3)
print(m2, m2.shape)
```

```python
m1 = np.arange(12)

m3 = m1.reshape(2,2,3)
print(m3, m3.shape)
```

```python
m1 = np.arange(6)

m2 = m1. reshape(-1,2)
print(m2, m2.shape)

m2 = m1.reshape(-1, 3)
print(m2, m2.shape)

m2 = m1.reshape(-1, 1)
print(m2, m2.shape)
```

```python
m1 = np.arange(6)

m3 = m1. reshape(-1,2,3)
print(m3, m3.shape)

m3 = m1.reshape(-1, 3, 2)
print(m3, m3.shape)

m3 = m1.reshape(2,3,-1)
print(m3, m3.shape)
```

### 배열의 통합과 분할

```python
a = np.arange(9).reshape(3,3)

b = a * 2
print(a)
print(b)
```

```python
r = np.vstack((a,b))
print(r, r.shape)
```

```python
c = np.row_stack((a,b))
print(c, c.shape)
```

```python
x = np.concatenate((a,b), axis = 0)
print(x, x.shape)
```

```python
import random
```

```python
r = np.hstack((a, b))
print(r, r.shape)
```

```python
c = np.column_stack((a, b))
print(c, c.shape)
```

```python
x = np.concatenate((a, b), axis = 1)
print(x, x.shape)
```
