### 1차원 배열의 인덱싱과 슬라이싱

```python
a = np.array([[1,2,3],
              [4,5,6]])

b = 2
print(a + b)
print(a - b)
print(a / b)
print(a * b)
```

```python
a = np.array([[1,2,3],
              [4,5,6]])

b = np.array([1,2,3])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
a = np.array([[1,2],
              [3,4]])

b = np.array([[1,2],
              [1,2]])

print(a * b) # 배열의 곱셈
print(a.dot(b)) #행렬의 곱셈
```
