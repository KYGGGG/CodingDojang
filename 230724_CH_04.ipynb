###  클래스 사용

```python
# 클래스 : 설계도 (붕어빵 틀)
# 객체 : 실제 자동차 (붕어빵)
# 인스턴스 : 인스턴트 (통조림) -> 일회용
# __init__ -> 자동 호출
```

```python
# def __init__(self)
# self -> 나 자신
class Stocks : # 선언
  def __init__(self, name, price) :
    self.name = name
    self.price = price

lg = Stocks("LG전자", 150000) # 호출
print(lg.name, lg.price)

```

```python
class Stocks :
  def __init__(self, name, price, count = 0) :
    self.name = name
    self.price = price
    self.count = count

lg = Stocks("LG전자", 150000, 2)
print(lg.name, lg.price, lg.count)
```

```python
class Stocks :
  def __init__(self, name, price, count = 0) :
    self.name = name
    self.price = price
    self.count = count
  def click(self, num_times = 1) :
    self.num_times = num_times
    self.count = self.count + self.num_times

lg = Stocks("LG전자", 150000)
print(lg.name, lg.price, lg.count)
lg.click()
print(lg.name, lg.price, lg.count)
lg.click()
print(lg.name, lg.price, lg.count)
```

```python
class Stocks :
  def __init__(self, name, price, count = 0) :
    self.name = name
    self.price = price
    self.count = count

  def __repr__(self) :
    return f"Stock 클릭 수(count = {self.count})"

  def click(self, num_times = 1) :
    self.num_times = num_times
    self.count = self.count + self.num_times

  def read(self) :
    return self.count

  def reset(self) :
    self.count = 0


lg = Stocks("LG전자", 150000)
print(lg.name, lg.price, lg.count)
lg.click()
print(lg.name, lg.price, lg.count)
lg.click()
print(lg.name, lg.price, lg.count)
lg.click()
print(lg.name, lg.price, lg.count)
lg.read()
print(lg.name, lg.price, lg.count)
lg.reset()
print(lg.name, lg.price, lg.count)
lg.click()
lg.click()
lg.click()
lg.__repr__()
```

```python
class subStocks(Stocks) :
  def reset(self) :
    pass

  def subprice(self, price1) :
    self.price1 = price1
    return f"subprice 반환", self.price1

  def addcountprice(self, count, price) :
    self.price = price
    self.count = count
    return f"price + count 합", self.price + self.count

sublg = subStocks("LG전자", 150000)
print(sublg.name, sublg.price, sublg.count)

sublg.click()
sublg.click()
sublg.click()
print(sublg.name, sublg.price, sublg.count)

sublg.reset()
print(sublg.name, sublg.price, sublg.count)
```

```python
sublg.subprice(180000)
print(sublg.name, sublg.price, sublg.count, sublg.subprice(180000))
```

```python
sublg1= subStocks("LG전자", 150000)

sublg1.click()
sublg1.click()
print(sublg1)

sublg1.addcountprice(5, 150000)
```

```python
class subsubStocks(subStocks) :
  def reset(self) :
    self.count = 100

  def click(self, num_times = 2) :
    self.count = self.count * num_times

  def subsubprice(self, price1) :
    self.price1 = price1
    return f"subprice 반환", self.price1

  def addcountprice(self, count, price) :
    self.price = price
    self.count = count
    return f"price + 100 * count", self.price + 100 * self.count

sublg7 = subStocks("LG전자7", 140000)
subsublg7 = subsubStocks("LG전자7", 140000)
print(subsublg7)
subsublg7.reset()
subsublg7.click()
subsublg7.click()
subsublg7.click()
print(subsublg7)
```

```python
print(subsublg7.name, subsublg7.price, subsublg7.count)
print(sublg7.addcountprice(5, 150000))
print(subsublg7.addcountprice(5, 150000))
```
